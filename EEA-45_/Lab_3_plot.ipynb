```python
import numpy as np
import matplotlib.pyplot as plt

Vz=6.2
def v(X,vz):
    if(X<=-0.6-vz):
        return -0.6-vz
    if(-0.6-vz<X and X<=0.6+2*vz):
        return X
    if(X>0.6+2*vz):
        return 0.6+2*vz
x=np.linspace(-30,30,1000)
y=np.zeros(len(x))
for i in np.arange(0,len(x),1):
    y[i]=v(x[i],Vz)
plt.plot(x,y, color='lightblue', linewidth=3)
plt.xlabel(r"$V_{O}(V_{I})$")
plt.ylabel(r"$V_{I}$")
plt.show()
```
