```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
pi=np.pi
t=np.linspace(0.0, 2*pi, num=N)
w=1.0

x = 3*np.cos(w*t)
y = 3*np.cos(w*t)

plt.scatter(x, y)
plt.show()
```

```python
N = 200
pi=np.pi
t=np.linspace(0.0, 2*pi, num=N)
w=1.0

x = 3.0*np.cos(w*t)
y = -3.0*np.sin(w*t)

fig, axs = plt.subplots(1, 1)

"""
 Single plot, não há indexação
"""

#axs[0, 0].plot(x, y)
#axs[0, 0].set_aspect('equal', 'box')
#axs[0, 0].set_title('still a circle, auto-adjusted data limits', fontsize=10)

axs.plot(x, y)
axs.set_aspect('equal', 'box')
axs.set_title('still a circle, auto-adjusted data limits', fontsize=10)


fig.tight_layout()

plt.show()
```
