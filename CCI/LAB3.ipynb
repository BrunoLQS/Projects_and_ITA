<p style=”text-align: justify;”>&nbsp;&nbsp;Com o intuito de comparar a qualidade das regressões, será utilizado o coeficiente de determinação $r$, também conhecido como coeficiente de correlação linear. Esse coeficiente mede o grau de relacionamento linear entre valores emparelhados $x$ e $y$ em uma amostra.&nbsp;Nesse sentido:<br />
$r^2=1-\frac{\sum_{i=1}^{n}e_i^2}{\sum_{i=1}^{n}(y_i-\bar{y})^2}$<bar />,
onde $e_i= y_i - Y_i$ é o resíduo da i-ésima observação. 
Assim, a qualidade da regressão será tão melhor quanto maior for essa medida.
</p> 

```python
def f1(x1,x2,x3,x4):
    y= a1[0]*x1+a1[1]*x2+a1[2]*x3+a1[3]*x4
    return y

def f2(x1,x2,x3,x4):
    y= a2[0]*x1+a2[1]*x2+a2[2]*x3+a2[3]*x4
    return y

def f3(x1,x2,x3,x4):
    y= a3[0]*x1+a3[1]*x2+a3[2]*x3+a3[3]*x4
    return y
```

```python
Y1=f1(norte['pagos'],
      norte['carga.paga'],
      norte['carga.gratis'],
      norte['correio']
     )

Y2=f2(sul['pagos'],
      sul['carga.paga'],
      sul['carga.gratis'],
      sul['correio']
     )

Y3=f3(eur['pagos'],
      eur['carga.paga'],
      eur['carga.gratis'],
      eur['correio']
     )
```

```python
# para y1 (NORTE):
V1=(y1-Y1)**2
n1=V1.sum()
average=np.average(y1)
D1=(y1-average)**2
d1=D1.sum()
r1=1-(n1/d1)
print("O fator de qualidade referente a norte é: %.4f" %r1)

# para y2  (SUL):
V1=(y2-Y2)**2
n1=V1.sum()
average=np.average(y2)
D1=(y2-average)**2
d1=D1.sum()
r2=1-(n1/d1)
print("O fator de qualidade referente a sul é: %.4f" %r2)

# para y3 (EUR):
V1=(y3-Y3)**2
n1=V1.sum()
average=np.average(y3)
D1=(y3-average)**2
d1=D1.sum()
r3=1-(n1/d1)
print("O fator de qualidade referente a eur é: %.4f" %r3)

print("\n\nPortanto,a regressão referente ao destino norte tem maior qualidade")
print("para explicar os passageiros não pagantes para o destino.")
```

<p style="text-align: justify;">&nbsp;&nbsp;Em geral, o número $n$ de valores das grandezas variáveis é maior que o número $m$ de incógnitas e devido a erros intrínsecos à aquisição dos dados e erros aleatórios, assim, o sistema linear é inconsistente e só pode ser resolvido de forma aproximada.&nbsp;Desse modo, surge uma dificuldade em determinar precisamente a família de funções da qual se aproximam as correlações entre as variáveis.
</p>

<p style="text-align: justify;">&nbsp;&nbsp;O Teorema da Aproximação de Weierstrass afirma que qualquer função contínua definida em um intervalo fechado [a,b] pode ser aproximada uniformemente tanto quanto se queira por uma função polinomial.
</p>

<p style="text-align: justify;">&nbsp;&nbsp;O Teorema da Aproximação de Weierstrass afirma que qualquer função contínua definida em um intervalo fechado [a,b] pode ser aproximada uniformemente tanto quanto se queira por uma função polinomial.<br / >
&nbsp;&nbsp;Desse modo, proponho o ajuste:<br />
$y_j(x_1,x_2,x_3,x_4)= \sum_{n=0}^{3}{a_nx_{1,j}^n+b_nx_{2,j}^n+c_nx_{3,j}^n+d_nx_{4,j}^n}$<br />
Isso é equivalente a:<br />
$y_j= (x_{1,1}^0,x_{1,1}^1,\dots,x_{1,1}^3,x_{2,1}^0,\dots,x_{3,1}^3)\cdot(a_0,a_1,\dots,d_2,d_3)^T$ <br />
Seja: <br />
$A_j=(x_{1,j}^0,x_{1,j}^1,\dots,x_{1,j}^3,x_{2,j}^0,\dots,x_{3,j}^3)$<br />
$X_j^T=(a_0,a_1,\dots,d_2,d_3)^T$<br />
Temos:<br />
$y_j=A_j \cdot X_j$<br />

<p style="text-align: justify;">&nbsp;&nbsp;Correlação e R-quadrado.<br/><br/>

&nbsp;&nbsp;No contexto de modelos preditórios (usualmente regressões lineares), onde $y$ é a verdade e $f$ é o modelo de predição, a definição desse fator é expresso por:<br /><br />

$r^2= 1 - \frac{\sum_{i=1}^{n}(y_i-f_i)^2}{\sum_{i=1}^{n}(y_i-\overline{y})^2}$<br /><br />
&nbsp;&nbsp;Em outras palavras, $r^2$ mede o tamanho da variância de $y$ sendo explicado pelo modelo, $f$. <br /><br />

$L(f,y)=\sum_{i=1}^{n}(f_i-y_i)^2 \leq\sum_{i=1}^{n}(af_i+b-y_i)^2$</p>

<p style="text-align: justify;">Seja $\displaystyle g(a,b)=\sum_{i=1}^{n}(af_i+b-y_i)^2$<br />
A melhor otimização é para $g(1,0)$$\Rightarrow$<br />
$\displaystyle \frac{\partial}{\partial a}g(a,b)_{(a=1,b=0)}=\sum_{i=2}^{n}2(af_i +b -y_i)f_i$ <br />
$\displaystyle \frac{\partial}{\partial a}g(a,b)_{(a=1,b=0)}=\sum_{i=2}^{n}2(f_i-y_i)f_i $<br />
$\displaystyle=0$<br />
$\displaystyle \frac{\partial}{\partial b}g(a,b)_{(a=1,b=0)}=\sum_{i=2}^{n}2(af_i +b -y_i)$ <br />
$\displaystyle=\sum_{i=2}^{n}2(f_i -y_i)$<br />
$\displaystyle =0$<br />
Da parcial com respeito a a, nós obtém-se:
$\displaystyle f\cdot f = y\cdot f$<br />
e da parcial com respeito a b, obtém-se:
$\displaystyle \overline{f}=\overline{y}$<br />
(desde que isso seja exatamente a soma normalizada)<br />
Agora,façamos uma mudança de variável no sistema de coordenadas de modo que $\overline{y}$( e $\overline{f}$ ) sejam iguais a zero. Isso faz as equações ficarem muito simples e não afetam a generalidade do resultado:<br />
A equação para $ r^2 $ é agora: <br />
$r^2=1-\frac{\sum_{i=1}^{n}(y_i-f_i)^2}{\sum_{i=1}^{n}y_i^2}$<br />
$\,\,\,\,\,\,= 1 - \frac{y\cdot y-2y\cdot f + f\cdot f}{y \cdot y}$ <br />
$\,\,\,\,\,\,=1-\frac{y\cdot y - f\cdot f}{y \cdot y}$<br />
$\,\,\,\,\,\,= \frac{f \cdot f}{y \cdot y}$<br />
A equação para correlação é : <br />
$\rho \,\,= \frac{\frac{1}{n}\sum_{i=1}^{n}f_iy_i}{\sqrt{(\frac{1}{n}\sum_{i=1}^{n}f_i^2)(\frac{1}{n}\sum_{i=1}^{n}y_i^2)}}$<br />
$\,\,\,\,\,\,= \frac{f \cdot y}{\sqrt{(f \cdot f)(y \cdot y)}}$<br />
$\,\,\,\,\,\,=\frac{f \cdot f}{\sqrt{(f \cdot f)(y \cdot y)}}$<br />
$\,\,\,\,\,\,=\sqrt{ \frac{f \cdot f}{y \cdot y} }$<br />
Portanto, $\rho^2 = r^2$.

$ S'_{ii} = c^2\, S_{ii}  -  2\, s c \,S_{ij}  +  s^2\, S_{jj}$ <br />

$ S'_{ik} = S'_{ki} = c \, S_{ik}  -  s \, S_{jk} \,\,\,\,\,\,
k \neq i,j $ <br />
$ S'_{jk} = S'_{kj} = s \, S_{ik}  + c \, S_{jk} ,\,\,\,\,\,
k \neq i,j $<br />
$ S'_{kl} = S_{kl} ,\,\,\,\,\,
k,l \neq i,j$ <br />
<br />
    
Onde $s=\sin(\theta)$ e $ c = \cos(\theta) $.<br /><br />
Como G é ortogonal, S e S′ têm a mesma norma de Frobenius , que é a raiz quadrada da soma dos quadrados de todos os elementos) , o que nos permite escolher $\theta$ que satisfaça $S_{ij}^{'}=0$, no caso em que $S^{'}$ tem a maior soma dos quadrados na diagonal:<br /><br />

$S'_{ij} = \cos(2\theta) S_{ij} + \tfrac{1}{2} \sin(2\theta) (S_{ii} - S_{jj})$

Igualando-se isso a 0 e rearranjando, temos:<br /><br />

$\tan(2\theta) = \frac{2 S_{ij}}{S_{jj} - S_{ii}}$<br /> <br />

se $ S_{jj} = S_{ii}$ (caso que a expressão acima não contempla), teremos: <br /> <br />

$\theta = \frac{\pi} {4} $ <br /><br />

O método de Jacobi para autovalores efetua repetidamente rotações até que a matriz se torne aproximadamente diagonal por meio do ataque ao pivô a cada iteração. Assim, os elementos finais na diagonal principal são aproximações dos autovalores reais de $S$.
