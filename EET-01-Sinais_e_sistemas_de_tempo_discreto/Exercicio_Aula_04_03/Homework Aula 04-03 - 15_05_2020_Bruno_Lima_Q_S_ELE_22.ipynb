### Propriedade 0

$X(e^{jw})$ é periódica e tem período $2\pi$

```python
From = -10
To = 10
Length = To - From + 1

# vetor de instantes
domain = np.linspace(start=From,stop=To, num = Length) # from and to inclusive.


# Step function
def u(n):
    if(n>=0):
        return 1.0
    else: 
        return 0
```

$$x[n]=a^{n}u[n] \;\;,\;\; |a|<1$$


Digamos $a=0.8$

```python
a=0.8
y=np.zeros(shape=(Length))
x=domain
```

```python
# x[n] <---> y
for i in np.arange(0,len(x),1):
    n=x[i]
    y[i]=u(n)*(a**n)  
    
# X(e^{jw}) <--> Y
##   w=10
  
w=10

Y=np.zeros(shape=(Length))

Y=Y.astype(complex)


for i in np.arange(0,len(domain),1):
    n=domain[i]
    Y[i]=y[i]*np.exp(-(w*n)*1j)
```

```python
Yp=np.zeros(shape=(Length))

Yp=Yp.astype(complex)


for i in np.arange(0,len(domain),1):
    n=domain[i]
    Yp[i]=y[i]*np.exp(-((w+2*np.pi)*n)*1j)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=2)

#plt.sca(axes[0, 0])   
plt.sca(axes[0,0])

markerline, stemlines, baseline=plt.stem(x,
                                         Y.real,
                                         markerfmt='or',
                                         linefmt='-r',
                                         basefmt=" ")
markerline.set_markerfacecolor('none')
plt.title('Propriedade 0')
plt.ylabel('Re(X(e^{10j}))')
plt.xlabel('n \n Figura(1): Parte Real de TFTD{x[n]}')
plt.xticks(np.arange(-10, 11, step=4))

plt.sca(axes[0,1])

markerline, stemlines, baseline=plt.stem(x,
                                         Y.imag,
                                         markerfmt='og',
                                         linefmt='-g',
                                         basefmt=" ")
markerline.set_markerfacecolor('none')
plt.title('Propriedade 0')
plt.ylabel('Img(X(e^{10j}))')
plt.xlabel('n \n Figura(2): Parte Imaginária de TFTD{x[n]}')
plt.xticks(np.arange(-10, 11, step=4))


plt.sca(axes[1,0])

markerline, stemlines, baseline=plt.stem(x,
                                         Yp.real,
                                         markerfmt='or',
                                         linefmt='-r',
                                         basefmt=" ")
markerline.set_markerfacecolor('none')
plt.title('Propriedade 0')
plt.ylabel('Re(X(e^{(10+2pi)j}))')
plt.xlabel('n \n Figura(3): Parte Real de TFTD{x[n]} \n com frequência deslocada em 2pi')
plt.xticks(np.arange(-10, 11, step=4))

plt.sca(axes[1,1])

markerline, stemlines, baseline=plt.stem(x,
                                         Yp.imag,
                                         markerfmt='og',
                                         linefmt='-g',
                                         basefmt=" ")
markerline.set_markerfacecolor('none')
plt.title('Propriedade 0')
plt.ylabel('Img(X(e^{(10+2pi)j}))')
plt.xlabel('n \n Figura(4): Parte Imaginária de TFTD{x[n]} \n com frequência deslocada em 2pi')
plt.xticks(np.arange(-10, 11, step=4))

#plt.show()
figure.tight_layout()
#plt.show()
```
