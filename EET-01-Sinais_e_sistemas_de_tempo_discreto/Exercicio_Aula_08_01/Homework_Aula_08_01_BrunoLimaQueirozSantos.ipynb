```python
from scipy.io import wavfile
fs, data = wavfile.read('blackbird.wav')
```

```python
fs
```

```python
samplerate=fs
length = data.shape[0] / samplerate
print(f"length = {length}s")
```

```python
import matplotlib.pyplot as plt
import numpy as np
time = np.linspace(0., length, data.shape[0])
plt.plot(time, data[:, 0], label="Left channel")
plt.plot(time, data[:, 1], label="Right channel")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Valor")
plt.show()
```

```python
from numpy.fft import fft
```

```python
fs, data = wavfile.read('blackbird.wav')
samplerate=fs
length = data.shape[0] / samplerate

x1=data[:, 0]
X1=fft(X1)
indexes=np.arange(1,len(X1)+1,1)


figure, axes = plt.subplots(nrows=1, ncols=2)

#plt.sca(axes[0,0])

plt.sca(axes[0])

plt.plot(indexes, X1.real, label="Left channel", color='r')
plt.title('X1=DFT{x1}')
plt.ylabel('Re(X1)')
plt.xlabel('n \n Figura(1): Parte Real de DFT{x[n]}')

#plt.sca(axes[0,0])
plt.sca(axes[1])

plt.plot(indexes, X1.imag, label="Left channel",color='green')
plt.title('X1=DFT{x1}')
plt.ylabel('Imag(X1)')
plt.xlabel('n \n Figura(2): Parte Imaginria de DFT{x[n]}')


figure.tight_layout()
```

Qual  a frequncia mxima do sinal x[n]?

A frequncia mxima do sinal x[n]  o inverso de seu perodo mnimo, que  5s.
Portanto, a frequncia mxima  0.2 Hz.

Qual deve ser a frequncia de amostragem mnima para assegurar que o sinal possa ser recuperado posteriormente por um processo de filtragem em banda base?

Pelo Teorema da Amostragem de Shannon, a frequncia mnima para assegurar que o sinal possa ser recuperado posteriormente por um processo de filtragem em banda base  0.5 Hz, maior que o dobro da frequncia do sinal.
