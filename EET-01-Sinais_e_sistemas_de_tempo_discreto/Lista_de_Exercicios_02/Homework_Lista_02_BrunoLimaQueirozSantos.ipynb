```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=np.arange(0,np.pi,0.01)
```

```python
def H(w):
  H=[]
  
  #H[0]=np.exp(-1j*2*w)
  H.append(np.exp(-1j*2*w))
  H[0]=H[0]/(1-(3/4)*np.exp(-1j*w))
  #H_1=np.exp(-1j*2*w)
  #H_1=H_1/(1-(3/4)*np.exp(-1j*w))
  #H[1]=-4*np.exp(-2j*w)/(4**2)
  H.append(-4*np.exp(-2j*w)/(4**2))
  H[1]=H[1]/(1-np.exp(-1j*w)/4)
  #H_2=-4*np.exp(-2j*w)/(4**2)
  #H_2=H_2/(1-np.exp(-1j*w)/4)
  #H[2]=np.exp(-1j*w)
  H.append(np.exp(-1j*w))
  return(H[0]+H[1]+H[2])
```

```python
h=H(w)
```

```python
def module(z):
  return(abs(z))

def argument(z):
  return(np.angle(z))
```

```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,5))
plt.sca(axes[0])

plt.plot(w, module(h), label="Magnitude", color='r')

plt.title('Magnitude Y(exp(jw))',fontsize=15)
plt.ylabel('|(Y(e^{jw})|',fontsize=15)
plt.xlabel('w \n Figura(1): Magnitude de TFTD{x[n]},',fontsize=15)
#plt.xticks(np.arange(-10, 11, step=4))



plt.sca(axes[1])

plt.plot(w,argument(h), label="Fase", color='green')

plt.title('Fase Y(exp(jw))',fontsize=15)
plt.ylabel('Fase(X(e^{wj}))',fontsize=15)
plt.xlabel('w \n Figura(1): Fase de TFTD{x[n]}',fontsize=15)
#plt.xticks(np.arange(-10, 11, step=4))


figure.tight_layout()
```
